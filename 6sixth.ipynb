```python
text_colors = ["Description", "Emotion", "Personality", "Mood", "Symbolism", "Use Case", "Keywords"]
dataset["Combined Text"] = dataset[text_colors].fillna("").agg(" ".join, axis=1)
dataset.head()
```

```python
dataset["Combined Text"][0]
```

```python
import re
def clean_text(text):
    text = str(text).lower().strip()
    text = re.sub(r'[^a-z0-9\s]+', " ", text)
    text = re.sub(r'\s', " ", text)
    return text.strip()
```

```python
dataset["Combined Text"] = dataset["Combined Text"].apply(clean_text)
dataset["Combined Text"][0]
```

```python
x = dataset["Combined Text"]
y = dataset[["R", "G", "B"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
print(type(x_train))
```

```python
import re

def remove_numbers(s):
    s = str(s)
    s = re.sub(r"\b\d+[a-zA-Z]*\b", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s
```

```python
x_train = x_train.apply(remove_numbers)
x_test  = x_test.apply(remove_numbers)
```

```python
vec = TfidfVectorizer(lowercase=True, stop_words="english", ngram_range=(1, 2), sublinear_tf=True, token_pattern=r"(?u)\b[a-zA-Z][a-zA-Z-]{1,}\b", min_df=5, max_df=0.9, max_features=20000)
x_train = vec.fit_transform(x_train)
x_test = vec.transform(x_test)
```

```python
feature_names = vec.get_feature_names_out()

# Print first 50 words (for quick check)
print(feature_names[:50])
feature_names.shape
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=300, max_features="sqrt", max_depth=30, min_samples_leaf=3, min_samples_split=3, max_samples=0.8, random_state=42, n_jobs=-1)
rf.fit(x_train, y_train)
```

```python
rf.score(x_test, y_test), rf.score(x_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "max_depth": [10, 20, 30, 40, 50],
    "min_samples_leaf":  [10, 20, 30, 40],
    "min_samples_split": [10, 20, 30], 
    "max_samples":       [0.6, 0.7, 0.8, 0.9], 
}
```

```python
grid = GridSearchCV(estimator=rf, param_grid=param_grid, scoring="r2", cv=3, verbose=2, refit=True)
grid.fit(x_train, y_train)
```

```python
print("Best Params:", grid.best_params_)
print("Best CV R2:", grid.best_score_)
```
