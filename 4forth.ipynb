```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_parquet("color_pedia.parquet")
dataset.head()
```

```python
dataset.dropna(inplace=True)
```

```python
def merge_text_cols(row):
    parts = [
        row["Description"],
        row["Emotion"],
        row["Personality"],
        row["Mood"],
        row["Symbolism"],
        row["Use Case"],
        row["Keywords"],
    ]

    parts = [str(p).strip() for p in parts if str(p).strip() not in ["", "nan", "None"]]
    return " ".join(parts)

dataset["Text"] = dataset.apply(merge_text_cols, axis=1)
dataset.head()
```
