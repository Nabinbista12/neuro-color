```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_parquet("color_pedia.parquet")
dataset.head()
```

```python
dataset = dataset.loc[0: 10000]
dataset.shape, dataset.size
```

```python
dataset.isnull().sum().sum()
```

```python
text_colors = ["Description", "Emotion", "Personality", "Mood", "Symbolism", "Use Case", "Keywords"]
dataset["Combined Text"] = dataset[text_colors].fillna("").agg(" ".join, axis=1)
dataset.head()
```

```python
import re
def clean_text(text):
    text = str(text).lower().strip()
    text = re.sub(r'[^a-z0-9\s]+', " ", text)
    text = re.sub(r'\s', " ", text)
    return text.strip()
```

```python
dataset["Combined Text"] = dataset["Combined Text"].apply(clean_text)
dataset["Combined Text"][0]
```

```python
x = dataset["Combined Text"]
y = dataset[["R", "G", "B"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import re

def remove_numbers(s):
    s = str(s)
    s = re.sub(r"\b\d+[a-zA-Z]*\b", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s
```

```python
x_train = x_train.apply(remove_numbers)
x_test  = x_test.apply(remove_numbers)
```

```python
vec = TfidfVectorizer(
    lowercase=True,
    stop_words="english",
    ngram_range=(1, 2),
    sublinear_tf=True,
    token_pattern=r"(?u)\b[a-zA-Z][a-zA-Z-]{1,}\b",
    min_df=5,
    max_df=0.9,
    max_features=20000,
)
x_train = vec.fit_transform(x_train)
x_test = vec.transform(x_test)
```

```python
from sklearn.svm import LinearSVR
from sklearn.multioutput import MultiOutputRegressor
```

```python
svr = MultiOutputRegressor(LinearSVR(random_state=42))
svr.fit(x_train, y_train)
```

```python
svr.score(x_test, y_test), svr.score(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
x_train.shape, x_test.shape
```

```python
rf = RandomForestRegressor(verbose=2)
rf.fit(x_train, y_train)
```

```python
rf.score(x_test, y_test), rf.score(x_train, y_train)
```
