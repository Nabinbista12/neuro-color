```python
text_colors = ["Description", "Emotion", "Personality", "Mood", "Symbolism", "Use Case", "Keywords"]
dataset["Combined Text"] = dataset[text_colors].fillna("").agg(" ".join, axis=1)
dataset.head()
```

```python
dataset["Combined Text"][0]
```

```python
import re
def clean_text(text):
    text = str(text).lower().strip()
    text = re.sub(r'[^a-z0-9\s]+', " ", text)
    text = re.sub(r'\s', " ", text)
    return text.strip()
```

```python
dataset["Combined Text"] = dataset["Combined Text"].apply(clean_text)
dataset["Combined Text"]
```

```python
x = dataset["Combined Text"]
y = dataset[["R", "G", "B"]]
```

```python
dataset["Combined Text"][0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vec = TfidfVectorizer(stop_words="english", ngram_range=(1, 2))
x_train = vec.fit_transform(x_train)
x_test = vec.transform(x_test)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, verbose=2)
rf.fit(x_train, y_train)
```

```python
rf.score(x_test, y_test), rf.score(x_train, y_train)
```

```python
text = vec.transform(["a vibrant and deep shade of maroon  passionate  intense bold  dramatic strong  powerful deep maroon often symbolizes power  strength  courage  and determination  it can also represent passion  desire  and deep emotions  ideal for creating a striking impact in design elements such as logos  branding  packaging  and advertising  powerful  passionate  bold  deep  strong  intense  maroon  ruby red"])
rf.predict(text)
```

```python
def predict_rf(text):
    text = vec.transform([text])
    rgb_predict = rf.predict(text)[0]
    rgb_int = np.clip(np.round(rgb_predict), 0, 255).astype(int)
    
    return rgb_int
```
