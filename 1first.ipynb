```python
text_colors = ["Description", "Emotion", "Personality", "Mood", "Symbolism", "Use Case", "Keywords"]
dataset["Combined Text"] = dataset[text_colors].fillna("").agg(" ".join, axis=1)
dataset.head()
```

```python
dataset["Combined Text"][0]
```

```python
import re
def clean_text(text):
    text = str(text).lower().strip()
    text = re.sub(r'[^a-z0-9\s]+', " ", text)
    text = re.sub(r'\s', " ", text)
    return text.strip()
```

```python
dataset["Combined Text"] = dataset["Combined Text"].apply(clean_text)
dataset["Combined Text"]
```

```python
x = dataset["Combined Text"]
y = dataset[["R", "G", "B"]]
```

```python
dataset["Combined Text"][0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vec = TfidfVectorizer(stop_words="english", ngram_range=(1, 2))
x_train = vec.fit_transform(x_train)
x_test = vec.transform(x_test)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.svm import LinearSVR
from sklearn.multioutput import MultiOutputRegressor
```

```python
svr = MultiOutputRegressor(LinearSVR(random_state=42))
svr.fit(x_train, y_train)
```

```python
svr.score(x_test, y_test), svr.score(x_train, y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test), dt.score(x_train, y_train)
```

```python
def predict_svr(text):
    text = vec.transform([text])
    rgb_predict = svr.predict(text)[0]
    rgb_int = np.clip(np.round(rgb_predict), 0, 255).astype(int)
    
    return rgb_int
```
