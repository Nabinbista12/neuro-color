```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_parquet("color_pedia.parquet")
dataset.head()
```

```python
dataset.shape, dataset.size
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape, dataset.size
```

```python
dataset[["Description", "Emotion",  "Personality",  "Mood", "Symbolism", "Use Case", "Keywords"]]
```

```python
print(dataset["Description"][0])
print(dataset["Emotion"][0])
print(dataset["Personality"][0])
print(dataset["Mood"][0])
print(dataset["Symbolism"][0])
print(dataset["Use Case"][0])
print(dataset["Keywords"][0])
```

```python
text_colors = ["Description", "Emotion", "Personality", "Mood", "Symbolism", "Use Case", "Keywords"]
dataset["Combined Text"] = dataset[text_colors].fillna("").agg(" ".join, axis=1)
dataset.head()
```

```python
dataset["Combined Text"][0]
```

```python
import re
def clean_text(text):
    text = str(text).lower().strip()
    text = re.sub(r'[^a-z0-9\s]+', " ", text)
    text = re.sub(r'\s', " ", text)
    return text.strip()
```

```python
dataset["Combined Text"] = dataset["Combined Text"].apply(clean_text)
dataset["Combined Text"][0]
```

```python
x = dataset["Combined Text"]
y = dataset[["R", "G", "B"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
print(type(x_train))
```

```python
import re

def remove_numbers(s):
    s = str(s)
    s = re.sub(r"\b\d+[a-zA-Z]*\b", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s
```

```python
x_train = x_train.apply(remove_numbers)
x_test  = x_test.apply(remove_numbers)
```

```python
vec = TfidfVectorizer(lowercase=True, stop_words="english", ngram_range=(1, 2), sublinear_tf=True, token_pattern=r"(?u)\b[a-zA-Z][a-zA-Z-]{1,}\b", min_df=5, max_df=0.9, max_features=20000)
x_train = vec.fit_transform(x_train)
x_test = vec.transform(x_test)
```

```python
feature_names = vec.get_feature_names_out()

# Print first 50 words (for quick check)
print(feature_names[:50])
feature_names.shape
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.svm import LinearSVR
from sklearn.multioutput import MultiOutputRegressor
```

```python
svr = MultiOutputRegressor(LinearSVR(random_state=42))
svr.fit(x_train, y_train)
```

```python
svr.score(x_test, y_test), svr.score(x_train, y_train)
```

```python
y_pred = svr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2_overall = r2_score(y_test, y_pred, multioutput="uniform_average")
mae_overall = mean_absolute_error(y_test, y_pred, multioutput="uniform_average")
mse_overall = mean_squared_error(y_test, y_pred, multioutput="uniform_average")
rmse_overall = np.sqrt(mse_overall)

print("R2  :", r2_overall)
print("MAE :", mae_overall)
print("MSE :", mse_overall)
print("RMSE:", rmse_overall)
```

```python
r2_rgb = r2_score(y_test, y_pred, multioutput="raw_values")
mae_rgb = mean_absolute_error(y_test, y_pred, multioutput="raw_values")
mse_rgb = mean_squared_error(y_test, y_pred, multioutput="raw_values")
rmse_rgb = np.sqrt(mse_rgb)

print("R2   [R,G,B]:", r2_rgb)
print("MAE  [R,G,B]:", mae_rgb)
print("MSE  [R,G,B]:", mse_rgb)
print("RMSE [R,G,B]:", rmse_rgb)
```

```python
def predict_svr(text):
    text = vec.transform([text])
    rgb_predict = svr.predict(text)[0]
    rgb_int = np.clip(np.round(rgb_predict), 0, 255).astype(int)
    
    return rgb_int
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = {
    "estimator__C": [0.1, 1.0, 10.0, 100.0],
    "estimator__epsilon": [0.0, 0.05, 0.1, 0.2],
    "estimator__loss": ["epsilon_insensitive", "squared_epsilon_insensitive"],
    "estimator__fit_intercept": [True, False],
}
```

```python
gs = GridSearchCV(estimator=svr, param_grid=param, verbose=2)
gs.fit(x_train, y_train)
```

```python
print("Best CV Score:", gs.best_score_)
print("Best Params:", gs.best_params_)
print("Best Model:", gs.best_estimator_)
```

```python
grid_svr = MultiOutputRegressor(LinearSVR(C=0.1, epsilon=0.05, loss="squared_epsilon_insensitive", fit_intercept=True, max_iter=20000, random_state=42))
grid_svr.fit(x_train, y_train)
```

```python
grid_svr.score(x_train, y_train), grid_svr.score(x_test, y_test)
```

```python
grid_predict = grid_svr.predict(x_test)
```

```python
r2_overall = r2_score(y_test, grid_predict, multioutput="uniform_average")
mae_overall = mean_absolute_error(y_test, grid_predict, multioutput="uniform_average")
mse_overall = mean_squared_error(y_test, grid_predict, multioutput="uniform_average")
rmse_overall = np.sqrt(mse_overall)

print("R2  :", r2_overall)
print("MAE :", mae_overall)
print("MSE :", mse_overall)
print("RMSE:", rmse_overall)
```

```python
r2_rgb = r2_score(y_test, grid_predict, multioutput="raw_values")
mae_rgb = mean_absolute_error(y_test, grid_predict, multioutput="raw_values")
mse_rgb = mean_squared_error(y_test, grid_predict, multioutput="raw_values")
rmse_rgb = np.sqrt(mse_rgb)

print("R2   [R,G,B]:", r2_rgb)
print("MAE  [R,G,B]:", mae_rgb)
print("MSE  [R,G,B]:", mse_rgb)
print("RMSE [R,G,B]:", rmse_rgb)
```
