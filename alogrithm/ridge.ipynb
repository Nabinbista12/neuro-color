```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_parquet("color_pedia.parquet")
dataset.head()
```

```python
dataset.shape, dataset.size, dataset.isnull().sum().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape, dataset.size
```

```python
text_colors = ["Description", "Emotion", "Personality", "Mood", "Symbolism", "Use Case", "Keywords"]
dataset["Combined Text"] = dataset[text_colors].fillna("").agg(" ".join, axis=1)
dataset.head()
```

```python
dataset["Combined Text"][0]
```

```python
import re
def clean_text(text):
    text = str(text).lower().strip()
    text = re.sub(r'[^a-z0-9\s]+', " ", text)
    text = re.sub(r'\s', " ", text)
    return text.strip()
```

```python
dataset["Combined Text"] = dataset["Combined Text"].apply(clean_text)
dataset["Combined Text"][0]
```

```python
x = dataset["Combined Text"]
y = dataset[["R", "G", "B"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import re

def remove_numbers(s):
    s = str(s)
    s = re.sub(r"\b\d+[a-zA-Z]*\b", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s
```

```python
x_train = x_train.apply(remove_numbers)
x_test  = x_test.apply(remove_numbers)
```

```python
vec = TfidfVectorizer(lowercase=True, stop_words="english", ngram_range=(1, 2), sublinear_tf=True, token_pattern=r"(?u)\b[a-zA-Z][a-zA-Z-]{1,}\b", min_df=5, max_df=0.9, max_features=20000)
x_train = vec.fit_transform(x_train)
x_test = vec.transform(x_test)
```

```python
feature_names = vec.get_feature_names_out()

print(feature_names[:50])
feature_names.shape
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import Ridge
```

```python
for x in range(20):
    rf = Ridge(alpha=x)
    rf.fit(x_train, y_train)
    print(x, rf.score(x_test, y_test), rf.score(x_train, y_train))
```

```python
ri = Ridge(alpha=9.0)
ri.fit(x_train, y_train)
```

```python
ri.score(x_test, y_test), ri.score(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2_overall = r2_score(y_test, y_pred, multioutput="uniform_average")
mae_overall = mean_absolute_error(y_test, y_pred, multioutput="uniform_average")
mse_overall = mean_squared_error(y_test, y_pred, multioutput="uniform_average")
rmse_overall = np.sqrt(mse_overall)

print("R2  :", r2_overall)
print("MAE :", mae_overall)
print("MSE :", mse_overall)
print("RMSE:", rmse_overall)
```

```python
r2_rgb = r2_score(y_test, y_pred, multioutput="raw_values")
mae_rgb = mean_absolute_error(y_test, y_pred, multioutput="raw_values")
mse_rgb = mean_squared_error(y_test, y_pred, multioutput="raw_values")
rmse_rgb = np.sqrt(mse_rgb)

print("R2   [R,G,B]:", r2_rgb)
print("MAE  [R,G,B]:", mae_rgb)
print("MSE  [R,G,B]:", mse_rgb)
print("RMSE [R,G,B]:", rmse_rgb)
```

```python
def predict_random_tree(mood):
    x_new = vec.transform([mood])         
    rgb_pred = rf.predict(x_new)[0]         
    rgb_int = np.clip(np.round(rgb_pred), 0, 255).astype(int)

    print("Mood:", mood)
    print("Predicted RGB (float):", rgb_pred)
    print("Predicted RGB (int 0-255):", rgb_int)
```

```python
mood = "strong powerful"
predict_random_tree(mood)
```
